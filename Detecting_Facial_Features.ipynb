# **Training**

**Importing libraries**

```python
from ultralytics import YOLO
```

**Importing Yolov8 Model**

```python
model = YOLO("yolov8n.pt")
```

**Model Training**

```python
results = model.train(
    data="/content/drive/MyDrive/Colab Notebooks/myDataset/data.yaml",
    epochs=100,
    imgsz=640,
    project="/content/drive/MyDrive/Colab Notebooks/Assessment/Detecting Facial Features/Custom Yolo ",
    name="gender_glasses_shirt_detection",
    exist_ok=True
)
```

**Importing model**

```python
model2 = YOLO("/content/drive/MyDrive/Colab Notebooks/Assessment/Detecting Facial Features/Custom Yolo /gender_glasses_shirt_detection/weights/best.pt")
```

# **Testing The model**

```python
import matplotlib.pyplot as plt
import cv2
import os

# Predict on the image
results = model2.predict("img5.jpg", save=True)

# Get the path of the saved prediction
predicted_image_path = os.path.join(results[0].save_dir, os.path.basename(results[0].path))

# Load and display the image
img = cv2.imread(predicted_image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.show()
```
