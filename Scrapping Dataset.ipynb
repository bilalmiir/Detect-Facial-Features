```
output_dir = "scraped_faces_dataset"
os.makedirs(output_dir, exist_ok=True)
```

```
TOTAL_IMAGES = 1000
images_per_query = TOTAL_IMAGES // len(search_queries) + 1

def download_images(query, max_images):
    query_folder = os.path.join(output_dir, query.replace(" ", "_"))
    os.makedirs(query_folder, exist_ok=True)

    downloaded = 0
    page = 1
    per_page = 15

    while downloaded < max_images:
        url = f"https://api.pexels.com/v1/search?query={query}&per_page={per_page}&page={page}"
        response = requests.get(url, headers=headers)

        if response.status_code != 200:
            print(f"Error fetching {query} page {page}: {response.status_code}")
            break

        data = response.json()
        photos = data.get('photos', [])
        if not photos:
            print(f"No more photos found for {query}.")
            break

        for photo in photos:
            if downloaded >= max_images:
                break
            img_url = photo['src']['medium']
            img_id = photo['id']
            try:
                img_data = requests.get(img_url).content
                filename = f"{query.replace(' ', '_')}_{img_id}.jpg"
                with open(os.path.join(query_folder, filename), 'wb') as f:
                    f.write(img_data)
                downloaded += 1
            except Exception as e:
                print(f"Failed to download {img_url}: {e}")
        
        page += 1
```

## Performing Scraping

```
for query in search_queries2:
    download_images(query, max_images=images_per_query)
```
